```python
#Fill missing values in 'Precip Type' with "none"
df['Precip Type'].fillna('none', inplace=True)
```

```python
#Encode Precip Type into numeric labels (LabelEncoder).
label_encoder = LabelEncoder()
df['Precip Type'] = label_encoder.fit_transform(df['Precip Type'])
df['Summary'] = label_encoder.fit_transform(df['Summary'])

print("\nUnique values in 'Precip Type' after encoding:\n", df['Precip Type'].unique())
```

```python
# Check if the columns exist before dropping them
columns_to_drop = ['Formatted Date', 'Daily Summary', 'Summary']
columns_in_df = df.columns

# Drop only the columns that exist in the dataframe
columns_to_drop = [col for col in columns_to_drop if col in columns_in_df]
df.drop(columns=columns_to_drop, inplace=True)

# Define X and y
X = df.drop(columns=['Precip Type'])  # Features: All columns except 'Precip Type'
y = df['Precip Type']  # Target: 'Precip Type'
```

```python
#Scaling
numeric_columns = X.select_dtypes(include=['float64', 'int64']).columns  # Select numeric columns
scaler = StandardScaler()
X[numeric_columns] = scaler.fit_transform(X[numeric_columns])
```

# Machine Learning

```python
#Splitting the data using train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% for testing
    random_state=42,    # ensures reproducibility
    shuffle=True        # shuffle before splitting (default=True)
)

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape:", X_test.shape, y_test.shape)
```

```python
# Check if there are any remaining missing values in numeric columns
print("\nMissing Values in Numeric Columns:\n", X.isnull().sum())

# Ensure all columns are numeric after encoding
print("\nData Types after Encoding:\n", X.dtypes)
#Train RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)
```

1- Compare ensemble methods (Random Forest vs Gradient Boosting vs XGBoost).

2- Which one would you recommend for real-world weather prediction, and why?

1 - Random Forest, Gradient Boosting, and XGBoost are all ensemble methods, which means they use many decision trees together to make better predictions.

Random Forest builds many trees at the same time (in parallel) and combines their results. It is simple, stable, and less likely to overfit, but sometimes it is not the most accurate.

Gradient Boosting builds trees one after another. Each new tree tries to fix the mistakes of the previous trees. This makes it more powerful than Random Forest, but also slower and more sensitive to overfitting.

XGBoost (Extreme Gradient Boosting) is an improved version of Gradient Boosting. It is much faster, uses memory efficiently, and has extra tools to prevent overfitting. It is widely used in machine learning competitions because of its high accuracy and speed

\2- For real-world weather prediction, I would recommend XGBoost. Weather data is usually very large and complex, and XGBoost can handle this efficiently. It is faster than Gradient Boosting and usually gives more accurate results than Random Forest. Also, XGBoost has good ways to prevent overfitting, which is important because weather patterns can be noisy and hard to predict.
